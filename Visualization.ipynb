## 数据转换可视化

从直方图和箱线图中可以看出，可用额度比值、负债率、月收入三个属性的分布极不均匀。其中可用额度比值、负债率主要分布在0-1之间，但最大值可以达到10^4，分布极不均匀，月收入也主要集中在10^3-10^4，但最大可以达到10^6。考虑对这三个属性取对数，取对数前后箱图对比以及对数值分布的直方图如图

```python
data['LogDebtRatio'] = data['DebtRatio'].apply(
    lambda x: math.log(x) if x else np.nan) #取对数，下同
plt.subplots(figsize=(8, 5)) #图像大小设置
plt.subplot(121) #子图排列
data.boxplot(column='DebtRatio') #取对数前的箱线图
plt.xticks([1], ['Original'], fontproperties='Times New Roman', size=14) #横轴标签设置和字体设置
plt.yticks(fontproperties='Times New Roman', size=14)
plt.subplot(122)
data.boxplot(column='LogDebtRatio') #取对数后的箱线图
plt.xticks([1], ['Logarithmic'], fontproperties='Times New Roman', size=14)
plt.yticks(fontproperties='Times New Roman', size=14)
plt.suptitle('Before and After Logarithmic Comparison\n(DebtRatio)',
             fontproperties='Times New Roman', size=16)
#plt.savefig('007_DebtRatio_box_Logarithmic_Comparison.png', dpi=1200) #保存到本地
plt.show()

data['LogRevolvingUtilizationOfUnsecuredLines'] = data['RevolvingUtilizationOfUnsecuredLines'].apply(
    lambda x: math.log(x) if x else np.nan)
plt.subplots(figsize=(8, 5))
plt.subplot(121)
data.boxplot(column='RevolvingUtilizationOfUnsecuredLines')
plt.xticks([1], ['Original'], fontproperties='Times New Roman', size=14)
plt.yticks(fontproperties='Times New Roman', size=14)
plt.subplot(122)
data.boxplot(column='LogRevolvingUtilizationOfUnsecuredLines')
plt.xticks([1], ['Logarithmic'], fontproperties='Times New Roman', size=14)
plt.yticks(fontproperties='Times New Roman', size=14)
plt.suptitle('Before and After Logarithmic Comparison\n(RevolvingUtilizationOfUnsecuredLines)',
             fontproperties='Times New Roman', size=16)
#plt.savefig('008_RUOUL_box_Logarithmic_Comparison.png', dpi=1200)
plt.show()

data['LogMonthlyIncome'] = data['MonthlyIncome'].apply(
    lambda x: math.log(x) if x else np.nan)
plt.subplots(figsize=(8, 5))
plt.subplot(121)
data.boxplot(column='MonthlyIncome')
plt.xticks([1], ['Original'], fontproperties='Times New Roman', size=14)
plt.yticks(fontproperties='Times New Roman', size=14)
plt.subplot(122)
data.boxplot(column='LogMonthlyIncome')
plt.xticks([1], ['Logarithmic'], fontproperties='Times New Roman', size=14)
plt.yticks(fontproperties='Times New Roman', size=14)
plt.suptitle('Before and After Logarithmic Comparison\n(MonthlyIncome)',
             fontproperties='Times New Roman', size=16)
#plt.savefig('009_MonthlyIncome_box_Logarithmic_Comparison.png', dpi=1200)
plt.show()
```

## 特征增加可视化

对月收入手动分箱，月收入小于180的认为是低收入

```python
#分箱
data['LowIncome'] = data['MonthlyIncome'] < 180
data['NormalIncome'] = data['MonthlyIncome'] >= 180 
data[['LowIncome', 'NormalIncome']].sum().plot(kind='bar') #作柱状图
plt.xticks(rotation=0, fontproperties='Times New Roman', size=14) #字体格式设置
plt.yticks(fontproperties='Times New Roman', size=14)
plt.title('MonthlyIncome',
          fontproperties='Times New Roman', size=16) #标题设置
#plt.savefig('013_Income_binning.png', dpi=1200) #保存到本地
plt.show()
```

对年龄手动分箱，年龄小于24的认为是青年，大于65的认为是老年

```python
#分箱
data['YoungAge'] = data['age'] < 24
data['OldAge'] = data['age'] > 65
data['NormalAge'] = (data['age'] <= 65) & (data['age'] >= 24)
data[['YoungAge', 'NormalAge', 'OldAge']].sum().plot(kind='bar') #作柱状图
plt.xticks(rotation=0, fontproperties='Times New Roman', size=14) #字体格式设置
plt.yticks(fontproperties='Times New Roman', size=14)
plt.title('Age',
          fontproperties='Times New Roman', size=16) #标题设置
#plt.savefig('014_Age_binning.png', dpi=1200) #保存到本地
plt.show()
```

增加维度：月收入的对数对家庭成员数取平均，对新特征作直方图

```python
data['LogIncomePerPerson'] = data['LogMonthlyIncome'] / \
    data['NumberOfDependents'] #增加维度
data.loc[~np.isfinite(data['LogIncomePerPerson']),
         'LogIncomePerPerson'] = np.nan 
data['LogIncomePerPerson'].hist(bins=100) #作直方图
plt.title('LogIncomePerPerson',
          fontproperties='Times New Roman', size=16) #设置标题
plt.xticks(fontproperties='Times New Roman', size=14) #字体格式设置
plt.yticks(fontproperties='Times New Roman', size=14)
#plt.savefig('015_LogIncomePerPerson.png', dpi=1200) #保存到本地
plt.show()
```

增加维度：月收入的对数乘以负债率的积再取对数，对新特征作直方图

```python
data['LogDebt'] = np.log(data['DebtRatio'] * data['LogMonthlyIncome']) #增加维度
data.loc[~np.isfinite(data['LogDebt']), 'LogDebt'] = np.nan
data['LogDebt'].hist(bins=100) #作直方图
plt.title('LogDebt',
          fontproperties='Times New Roman', size=16) #设置标题
plt.xticks(fontproperties='Times New Roman', size=14) #字体格式设置
plt.yticks(fontproperties='Times New Roman', size=14)
#plt.savefig('016_LogDebt.png', dpi=1200) #保存到本地
plt.show()
```

## 相关关系可视化

用热图表示各属性之间的相关关系，增加特征前与增加特征后的热图作对比

```python
original_data = pd.read_csv('cs-training.csv', index_col=0) #重新导入
original_data = original_data[original_data['age'] != 0]
original_data = original_data[original_data['NumberOfTime30-59DaysPastDueNotWorse'] < 80] #去除异常值
plt.subplots(figsize=(15, 15)) #图像大小设置
seaborn.heatmap(original_data.corr(), annot=True,
                vmax=1, square=True, cmap='Blues') #作相关系数热图
plt.title('Heatmap', size=20)
#plt.savefig('017_Heatmap.png', dpi=1200,
#            bbox_inches=trs.Bbox([[-2, -1], [13, 14]])) #保存到本地
plt.show()

#增加的所有特征
def add_features(data):
    data['NoDependents'] = data['NumberOfDependents'] == 0
    data['NumberOfDependents'].fillna(0, inplace=True)

    data['NoIncome'] = data['MonthlyIncome'] == 0
    data['LogIncome'] = np.log(data['MonthlyIncome'])
    data.loc[~np.isfinite(data['LogIncome']), 'LogIncome'] = 0
    data.loc[pd.isnull(data['LogIncome']), 'LogIncome'] = 0
    
    data['LowIncome'] = data['MonthlyIncome'] < 180
    data['YoungAge'] = data['age'] < 24
    data['OldAge'] = data['age'] > 65
    
    data['LogAge'] = np.log(data['age'])

    data['LogIncomePerPerson'] = data['LogIncome'] / data['NumberOfDependents']
    data.loc[~np.isfinite(data['LogIncomePerPerson']), 'LogIncomePerPerson'] = 0
    
    data['NoDebtRatio'] = data['DebtRatio'] == 0
    data['LogDebtRatio'] = np.log(data['DebtRatio'])
    data.loc[~np.isfinite(data['LogDebtRatio']), 'LogDebtRatio'] = 0
    data['LogDebt'] = np.log(data['DebtRatio'] * data['LogIncome'])
    data.loc[~np.isfinite(data['LogDebt']), 'LogDebt'] = 0
    
    data['LogDebtPerPerson'] = data['LogDebt'] / data['NumberOfDependents']
    data.loc[~np.isfinite(data['LogDebtPerPerson']), 'LogDebtPerPerson'] = 0

    data['NoRevolvingUtilizationOfUnsecuredLines'] = data['RevolvingUtilizationOfUnsecuredLines']==0
    data['LogRevolvingUtilizationOfUnsecuredLines'] = np.log(data['RevolvingUtilizationOfUnsecuredLines'])
    data.loc[~np.isfinite(data['LogRevolvingUtilizationOfUnsecuredLines']), 'LogRevolvingUtilizationOfUnsecuredLines'] = 0

    data.drop(['MonthlyIncome', 'age', 'RevolvingUtilizationOfUnsecuredLines', 'DebtRatio'], axis=1, inplace=True)
    
    return data

original_data = add_features(original_data) #增加特征
plt.subplots(figsize=(15, 15)) #图像大小设置
seaborn.heatmap(original_data.corr(), annot=True,
                vmax=1, square=True, cmap='Blues') #作增加特征后的相关系数热图
plt.title('Heatmap After Adding Features', size=20)
#plt.savefig('018_Heatmap_after_adding_features.png', dpi=1200,
#            bbox_inches=trs.Bbox([[-2, -1], [13, 14]])) #保存到本地
plt.show()
```
